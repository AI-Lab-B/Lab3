```python
## Q1 BFS

def valid_direction(matSize,cell):
    if (cell[0]<0 or cell [1] <0):
        return False
    if (cell[0]>matSize[0] or cell[1] > matSize[1] ):
        return False
    return True


def find_shortest_path(matrix):

    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    start = (0, 0)
    end = (3, 3)

    queue = []
    visited = []
    matRow=len(matrix)
    matCol=len(matrix[0])
    # BFS Loop
    queue.append((start,None))
    while queue:
        node, _ = queue.pop()
        if ( node == end ):
            return True
        if node not in visited:
            visited.append(node)
            for i in range(4):
                nodeCopy = [x for x in node]
                nodeCopy[0]=node[0]+directions[i][0]
                nodeCopy[1]=node[1]+directions[i][1]
                if (valid_direction((matRow,matCol),nodeCopy)):
                    queue.append(((nodeCopy[0],nodeCopy[1]),node))

    return None
size=4
matrix=[ [0 for i in range(size)] for j in range(size)]
check=find_shortest_path(matrix)
if check == True:
    print("Found")
elif check== None:
    print("No Solution Found")

```

```python
import time
import copy
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def prprint(matrix):
    for row in matrix:
        print(row)

def state_to_tuple(state):
    st_t=[]
    for cell in state:
        st_t.append(cell)
    st_t=tuple(st_t)
    return st_t

def tuple_to_state(matrix):
    st=""
    for cell in matrix:
        st+=cell
    return st

def matrix_to_tuple(matrix):
    st_t=[]
    for i in matrix:
        for j in i:
            st_t.append(j)
    st_t=tuple(st_t)
    return st_t

def tuple_to_matrix(state):
    mat=[]
    i=0
    mat_l=[]
    for j in state:
        mat_l.append(j)
        if i%3 == 2:
            mat.append(copy.deepcopy(mat_l))
            mat_l.clear()
        i+=1
    return mat;

def valid_move(cell):
    if (cell[0]<0 or cell[1]<0):
        return False
    if (cell[0]>2 or cell[1]>2):
        return False
    return True

def get_moves(matrix):
    moves=[]
    blank_index=[]
    for index_i,row in enumerate(matrix):
        for index_j,cell in enumerate(row):
            if matrix[index_i][index_j]=='0':
                blank_index=[index_i,index_j]
                break;

    for direction in directions:
        #cell_copy=copy.deepcopy(blank_index)
        row=blank_index[0]+direction[0]
        col=blank_index[1]+direction[1]

        if (valid_move((row,col))):
            moves.append(direction)
    return (moves,blank_index)

def dfs_search(state,goal_state,solution,visited):

    state_t=matrix_to_tuple(state)
    state_s=tuple_to_state(state_t)

    if state_s == goal_state:
        return solution

    visited.add(state_s)

    moves,blank_index = get_moves(state)

    for move in reversed(moves):
        child = copy.deepcopy(state)

        swap_index = (blank_index[0] + move[0] , blank_index[1] + move[1])
        temp = child[blank_index[0]][blank_index[1]]
        child[blank_index[0]][blank_index[1]] = child[swap_index[0]][swap_index[1]]
        child[swap_index[0]][swap_index[1]] = temp
        #prprint(child)


        state_f=matrix_to_tuple(child)
        state_g=tuple_to_state(state_f)

        print(move,state_g)
        if state_g not in visited:
            #print(state_g)
            visited.add(state_g)
            solution.append(state)
            return dfs_search(child,goal_state,solution,visited)




def dfs(start_state, goal_state):
    t_mat=tuple_to_matrix(start_state)
    return dfs_search(t_mat,goal_state,[],set())

def main():
    # 1 2 0
    # 3 4 5
    # 6 7 8
    start_state = '120345678'#input("Enter start State: ")
    goal_state = '012345678'#input("Enter goal State: ")
    start_tuple = state_to_tuple(start_state)
    #goal_tuple = state_to_tuple(goal_state)
    print("-----------------")
    print("DFS Algorithm")
    print("-----------------")
    start_time = time.time()
    solution_path = dfs(start_tuple, goal_state)
    end_time = time.time()
    if solution_path:
        print("Time taken:", end_time - start_time, "seconds")
        print("Path Cost:", len(solution_path))
        print("No of Nodes Visited:", len(solution_path) + 1)
        for state in solution_path:
            for row in state:
                print(' '.join(row))
                print("-----")
    else:
        print("No solution found.")

if __name__=="__main__":
    main()
```
