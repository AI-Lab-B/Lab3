```python
import time

def state_to_tuple(state):
    """Convert a string state to a tuple representation."""
    stateList = []

    for i in range(3):
        stateList.append([])
        for j in range(3):
            stateList[i].append(int(state[(i * 3) + j]))

    return stateList

def tuple_to_state(matrix):
    strState = ""

    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            strState += f"{matrix[i][j]}"

    return strState

def get_moves(matrix):
    emptyPos = []
    moveList = []

    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 0:
                emptyPos = [i, j]

    if emptyPos != []:
        if emptyPos[0] < 2:
            moveList.append("down")

        if emptyPos[0] > 0:
            moveList.append("up")

        if emptyPos[1] < 2:
            moveList.append("right")

        if emptyPos[1] > 0:
            moveList.append("left")

    return moveList

def dfs(start_state, goal_state):
    stack = [(start_state, [])]
    visited = set()

    while stack:
        current_state, path = stack.pop()
        current_tuple = tuple(tuple(row) for row in current_state)

        if current_tuple == tuple(tuple(row) for row in goal_state):
            return path

        if current_tuple in visited:
            continue
        visited.add(current_tuple)

        moves = get_moves(current_state)
        for move in moves:
            new_state = [row[:] for row in current_state]
            empty_row, empty_col = next((r, c) for r in range(3) for c in range(3) if new_state[r][c] == 0)

            if move == "up":
                if empty_row > 0:
                    new_state[empty_row][empty_col], new_state[empty_row - 1][empty_col] = new_state[empty_row - 1][empty_col], new_state[empty_row][empty_col]
            elif move == "down":
                if empty_row < 2:
                    new_state[empty_row][empty_col], new_state[empty_row + 1][empty_col] = new_state[empty_row + 1][empty_col], new_state[empty_row][empty_col]
            elif move == "left":
                if empty_col > 0:
                    new_state[empty_row][empty_col], new_state[empty_row][empty_col - 1] = new_state[empty_row][empty_col - 1], new_state[empty_row][empty_col]
            elif move == "right":
                if empty_col < 2:
                    new_state[empty_row][empty_col], new_state[empty_row][empty_col + 1] = new_state[empty_row][empty_col+1], new_state[empty_row][empty_col]

            stack.append((new_state, path + [new_state]))

    return None

def main():
    """Main function to take input and execute the DFS algorithm."""
    start_state = input("Enter start State: ")
    goal_state = input("Enter goal State: ")

    start_tuple = state_to_tuple(start_state)
    goal_tuple = state_to_tuple(goal_state)

    print("-----------------")
    print("DFS Algorithm")
    print("-----------------")
    start_time = time.time()
    solution_path = dfs(start_tuple, goal_tuple)
    end_time = time.time()

    if solution_path:
        print("Time taken:", end_time - start_time, "seconds")
        print("Path Cost:", len(solution_path))
        print("No of Nodes Visited:", len(solution_path) + 1)

        for state in solution_path:
            for row in state:
                print(' '.join(map(str, row)))
            print("-----")
    else:
        print("No solution found.")

if __name__ == "__main__":
    main()
```

```python
from collections import deque

class Graph:
    def __init__(self, adjacency_list):
        """Initializes the graph with an adjacency list."""
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        """Returns the neighbors of a given node."""
        return self.adjacency_list[v]

    def h(self, n):
        """Heuristic function: estimates the cost from node n to the goal."""
        H = {
            'The': 4,
            'cat': 3,
            'dog': 3,
            'runs': 2,
            'fast': 1
        }
        return H.get(n, float('inf'))

    def a_star_algorithm(self, start_node, stop_node):
        """Implements the A* search algorithm to find the optimal path."""
        open_list = set([start_node])
        closed_list = set([])

        g = {}
        g[start_node] = 0

        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            for v in open_list:
                if (n is None) or ((g[v] + self.h(v)) < (g[n] + self.h(n))):
                    n = v

            if n is None:
                print("Path does not exist!")
                return None

            if n == stop_node:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                reconst_path.append(start_node)
                reconst_path.reverse()

                print(f"Path found: {'-'.join(reconst_path)}")
                print(f"Total cost: {g[stop_node]}")
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)
            closed_list.add(n)

        print("Path does not exist!")
        return None

# Define the graph given in adjacency list
adjacency_list = {
    'The': [('cat', 2), ('dog', 1)],
    'cat': [('runs', 3)],
    'dog': [('runs', 3)],
    'runs': [('fast', 1)],
    'fast': []
}

graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('The', 'fast')
```
