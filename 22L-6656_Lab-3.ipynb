```python
#Q1

def find_shortest_path(matrix):
    # Directions: Up, Down, Left, Right
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    # Initialize starting and ending positions
    start = (1, 1)
    end = (4, 4)
    # Initialize data structures for BFS
    queue = deque([(start, [start])]) # Use deque for BFS
    visited = set() # Track visited positions
    # BFS Loop
    while queue:
    # Dequeue the next position
       current_pos, path = queue.popleft()
    # Check if HOME is reached
       if current_pos == end:
           return path
    # Mark the position as visited
       visited.add(current_pos)
    # Explore neighboring cells (non-diagonal and not obstacles)
       for direction in directions:
           new_pos = (current_pos[0] + direction[0], current_pos[1] + direction[1])
           if new_pos[0] < 0 or new_pos[0] >= len(matrix) or new_pos[1] < 0 or new_pos[1] >= len(matrix[0]):
               continue
           if new_pos in visited or matrix[new_pos[0]][new_pos[1]] == 1:
               continue
           queue.append((new_pos, path + [new_pos]))
    #Return the shortest path If found, else indicate no path
    return None

matrix = [[0]*6 for i in range(6)]

print(find_shortest_path(matrix))
```

```python
#Q2

def state_to_matrix(state):
    """Convert a string state to a matrix representation."""
    matrix = []
    for i in range(0, 3):
        row = []
        for j in range(0, 3):
            row.append(state[i * 3 + j])
        matrix.append(row)
    print("\nMatrix :", matrix)
    return matrix

def matrix_to_state(matrix):
    """Convert a matrix representation back to a string state."""
    state = ""
    for i in range(0, 3):
        for j in range(0, 3):
            state += matrix[i][j]
    return state
   
def get_moves(matrix):
    """Generate possible moves from the given state."""
    # Directions: Up, Down, Left, Right
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    moves = []

    # Find the empty tile (0)
    for i in range(0, 3):
        for j in range(0, 3):
            if matrix[i][j] == "0":
                x, y = i, j
                break

    # Generate all possible moves
    for direction in directions:
        new_x, new_y = x + direction[0], y + direction[1]
        if new_x >= 0 and new_x <=2 and new_y >= 0 and new_y <= 2:
            moves.append(direction)

    return moves

def dfs(start_matrix, goal_matrix):
    """Perform Depth-First Search (DFS) to find a solution path."""
    stack = [(start_matrix, [])]
    visited = set()
    while stack:
        current_matrix, path = stack.pop()
        state = matrix_to_state(current_matrix)
        print("State :", state)
        if state == matrix_to_state(goal_matrix):
            return path
        visited.add(state)
        for move in get_moves(current_matrix):
            new_matrix = [row[:] for row in current_matrix]
            x, y = 0, 0
            for i in range(0, 3):
                for j in range(0, 3):
                    if current_matrix[i][j] == "0":
                        x, y = i, j
                        break
            new_x, new_y = x + move[0], y + move[1]
            new_matrix[x][y], new_matrix[new_x][new_y] = new_matrix[new_x][new_y], new_matrix[x][y]
            new_state = matrix_to_state(new_matrix)
            if new_state not in visited:
                stack.append((new_matrix, path + [new_state]))
    return None
 
def main():
    """Main function to take input and execute the DFS algorithm."""
    start_state = "120345678"#input("Enter start State: ")
    goal_state = "012345678"#input("Enter goal State: ")
    start_matrix = state_to_matrix(start_state)
    goal_matrix = state_to_matrix(goal_state)
    print("-----------------")
    print("DFS Algorithm")
    print("-----------------")
    start_time = time.time()
    solution_path = dfs(start_matrix, goal_matrix)
    end_time = time.time()
    if solution_path:
        print("\nTime taken:", end_time - start_time, "seconds")
        print("Path Cost:", len(solution_path))
        print("No of Nodes Visited:", len(solution_path) + 1)
        for state in solution_path:
            state_to_matrix(state)
            print("-----")
    else:
        print("No solution found.") 

if __name__ == "__main__":
    main()
```

```python
#Q3 

class Graph:
    def __init__(self, adjacency_list):
        """Initializes the graph with an adjacency list."""
        self.adjacency_list = adjacency_list
 
    def get_neighbors(self, v):
        """Returns the neighbors of a given node."""
        # Implement the function to return neighbors of node v
        return self.adjacency_list[v]

    def h(self, n):
        """Heuristic function: estimates the cost from node n to the goal."""
        H = { # define the heuristics here
            'The': 4,
            'Cat': 3,
            'Dog': 3,
            'runs': 2,
            'fast': 1
        }
        return H[n]
    
    def a_star_algorithm(self, start_node, stop_node):
        """Implements the A* search algorithm to find the optimal path."""
        open_list = set([start_node])
        closed_list = set([])          

        g = {} # Cost from start node to all other nodes
        g[start_node] = 0          
        
        parents = {} # Keeps track of paths
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None # Current node

            # Find the node with the lowest f(n) = g(n) + h(n)
            
            for v in open_list:
                if n == None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n == None:
                print('Path does not exist!')
                return None
            
            # If goal node is found, reconstruct and return the path
                # print the path found

            if n == stop_node:
                path = []
                while parents[n] != n:
                    path.append(n)
                    n = parents[n]
                path.append(start_node)
                path.reverse()
                print('Path:', path)
                return path
            
            for (m, weight) in self.get_neighbors(n).items():
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)
            open_list.remove(n)
            closed_list.add(n)

       
# Define the graph given in adjacency list
adjacency_list = {
    'The': {'Cat': 2, 'Dog':3},
    'Cat': {'runs': 1},
    'Dog': {'runs': 2},
    'runs': {'fast': 2}
}
graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('The','fast') # Define start and goal words
```
