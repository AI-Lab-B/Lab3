```
import heapq
from collections import deque


class Matrix:
    def __init__(self, n=int(0)):
        self.matrix = list()
        self.size = int(n)
        for i in range(self.size):
            self.matrix.append([])
            for j in range(self.size):
                self.matrix[i].append(0)

    def display(self):
        for i in range(self.size):
            for j in range(self.size):
                print(self.matrix[i][j], end=" ")
            print("")

    def get_size(self):
        return self.size


def bfs(matrix=Matrix(), start=tuple(), goal=tuple()):

    queue = deque()
    visited = set()

    queue.append([start, "start_state"])

    while queue:
        element = queue.popleft()
        dimension = element[0]
        moves = element[1:]

        visited.add(dimension)
        x, y = dimension

        if dimension == goal:
            return moves

        if x > 0:
            new_dim = (x - 1, y)  # up
            if (new_dim not in queue) and (new_dim not in visited):
                queue.append([new_dim, *moves, "up"])
        if x < (matrix.get_size() - 1):
            new_dim = (x + 1, y)  # down
            if (new_dim not in queue) and (new_dim not in visited):
                queue.append([new_dim, *moves, "down"])
        if y > 0:
            new_dim = (x, y - 1)  # left
            if (new_dim not in queue) and (new_dim not in visited):
                queue.append([new_dim, *moves, "left"])
        if y < (matrix.get_size() - 1):
            new_dim = (x, y + 1)  # right
            if (new_dim not in queue) and (new_dim not in visited):
                queue.append([new_dim, *moves, "right"])

    return "not found"


def check(n, x_start, y_start, x_goal, y_goal):
    if (
        (x_start >= 0 and x_start <= n - 1)
        and (y_start >= 0 and y_start <= n - 1)
        and (y_goal >= 0 and y_goal <= n - 1)
        and (x_goal >= 0 and x_goal <= n - 1)
    ):
        return True


def main():
    n = int(input("Enter size of matrix"))
    x_start = int(input("Enter start x position "))
    y_start = int(input("Enter start y position "))
    x_goal = int(input("Enter goal x position "))
    y_goal = int(input("Enter goal y position "))
    if check(n, x_start, y_start, x_goal, y_goal):
        matrix = Matrix(n)
        print(bfs(matrix, (x_start, y_start), (x_goal, y_goal)))
    else:
        print("Wring co_ordinates")


if __name__ == "__main__":
    main()
```

```
from collections import deque
import copy


class Board:
    def __init__(self, state=str("")):
        self.board = list()
        state_arr = list(state)
        x = 0
        for i in range(3):  # rows
            self.board.append([])
            for j in range(3):  # columns
                self.board[i].append(int(state_arr[x]))
                x = x + 1

    def display(self):
        print("Matrix")
        for i in range(3):
            for j in range(3):
                print(self.board[i][j], end=" ")
            print("")

    def reached_goal_state(self, goal_board):
        if self.board == goal_board.board:
            return True

        return False

    def give_index_of_blank(self):
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == 0:
                    return (i, j)

    def move_blank_left(self):
        x, y = self.give_index_of_blank()
        if y != 0:
            new_x = x
            new_y = y - 1
            temp = self.board[new_x][new_y]

            self.board[new_x][new_y] = self.board[x][y]
            self.board[x][y] = temp

    def move_blank_right(self):
        x, y = self.give_index_of_blank()
        if y != (3 - 1):
            new_x = x
            new_y = y + 1
            temp = self.board[new_x][new_y]

            self.board[new_x][new_y] = self.board[x][y]
            self.board[x][y] = temp

    def move_blank_up(self):
        x, y = self.give_index_of_blank()
        if x != 0:
            new_x = x - 1
            new_y = y
            temp = self.board[new_x][new_y]

            self.board[new_x][new_y] = self.board[x][y]
            self.board[x][y] = temp

    def move_blank_down(self):
        x, y = self.give_index_of_blank()
        if x != (3 - 1):
            new_x = x + 1
            new_y = y
            temp = self.board[new_x][new_y]

            self.board[new_x][new_y] = self.board[x][y]
            self.board[x][y] = temp


def dfs(start_board, goal_board):
    queue = deque()
    visited = set()
    states_expanded = 0

    queue.append([start_board])

    while queue:
        boards = queue.pop()
        board = boards[-1]
        states_expanded = states_expanded + 1

        print("states_expanded: ", states_expanded)

        visited.add(board)

        if board.reached_goal_state(goal_board):
            return boards

        new_board_1 = copy.deepcopy(board)
        new_board_2 = copy.deepcopy(board)
        new_board_3 = copy.deepcopy(board)
        new_board_4 = copy.deepcopy(board)

        new_board_1.move_blank_down()
        new_board_2.move_blank_right()
        new_board_3.move_blank_up()
        new_board_4.move_blank_left()

        if new_board_1 not in visited:
            new_path = list(boards)
            new_path.append(new_board_1)
            queue.append(new_path)

        if new_board_2 not in visited:
            new_path = list(boards)
            new_path.append(new_board_2)
            queue.append(new_path)

        if new_board_3 not in visited:
            new_path = list(boards)
            new_path.append(new_board_3)
            queue.append(new_path)

        if new_board_4 not in visited:
            new_path = list(boards)
            new_path.append(new_board_4)
            queue.append(new_path)

    return "not found"


def check(start, goal):
    start_arr = list(start)
    goal_arr = list(goal)
    if len(start_arr) == 9 and len(goal_arr) == 9:
        return True

    return False


def main():
    start = input("Enter start state ")
    goal = input("Enter goal state ")
    if check(start, goal):
        start_board = Board(start)
        goal_board = Board(goal)
        start_board.display()
        goal_board.display()
        boards = dfs(start_board, goal_board)
        for board in boards:
            board.display()
    else:
        print("Wrong state")


if __name__ == "__main__":
    main()
```

```
from collections import deque
import heapq


class Graph:
    def __init__(self):
        self.graph = {}
        # key is a node and value is list of edges (to other node)

    def add_node(self, node):
        if node not in self.graph:
            self.graph[node] = []
        else:
            print(f"Node {node} already exists.")

    def add_edge(self, u, v, weight=1, directed=False):
        if (u in self.graph) and (v in self.graph):
            if directed:
                self.graph[u].append((v, weight))
                self.graph[v].append((u, weight))
            else:
                self.graph[u].append((v, weight))

        else:
            print(f"Error: One or both nodes ({u}, {v}) do not exist!")

    def print(self):
        for node in self.graph:
            print(f"{node} -> {self.graph[node]}")

    def get_neighbours(self, u):
        if u in self.graph:
            return self.graph[u]
        else:
            raise Exception("This nodes is not in graph")


def make_sentence_graph(graph=Graph()):
    graph.add_node("The")
    graph.add_node("cat")
    graph.add_node("dog")
    graph.add_node("runs")
    graph.add_node("fast")

    graph.add_edge("The", "cat", 2)
    graph.add_edge("The", "dog", 3)
    graph.add_edge("cat", "runs", 1)
    graph.add_edge("dog", "runs", 2)
    graph.add_edge("runs", "fast", 2)


def h(sentence_graph_node):
    h_dict = {"The": 4, "cat": 3, "dog": 3, "runs": 2, "fast": 1}
    return h_dict[sentence_graph_node]


def A_star(graph, start, goal):
    heap = [(h(start), 0, start, [start])]  # (f(n), g(n), current_node, path)
    heapq.heapify(heap)
    visited = set()

    while heap:
        f_n, g_n, current_node, path = heapq.heappop(heap)

        if current_node == goal:
            return path, g_n

        if current_node in visited:
            continue

        visited.add(current_node)

        for neighbour, weight in graph.get_neighbours(current_node):
            if neighbour not in visited:
                new_g_n = g_n + weight
                new_f_n = new_g_n + h(neighbour)
                heapq.heappush(heap, (new_f_n, new_g_n, neighbour, path + [neighbour]))

    return None, float("inf")


def main():
    g = Graph()
    make_sentence_graph(g)
    g.print()

    start_node = "The"
    goal_node = "fast"
    path, total_cost = A_star(g, start_node, goal_node)

    if path:
        print(f"Optimal path: {' -> '.join(path)}")
        print(f"Total cost: {total_cost}")
    else:
        print("No path found.")


if __name__ == "__main__":
    main()
```
