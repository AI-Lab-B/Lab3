```python
import matplotlib.pyplot as plt
import numpy as np
import random

with open("text.txt", "r") as file:
    genome_seq = file.read()

# Generate random genome sequence
nucleotides = ["A", "T", "C", "G"]
genome_length = len(genome_seq)
genome = "".join(random.choice(nucleotides) for _ in range(genome_length))

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(projection='3d')

for i, nucleotide in enumerate(genome):
    ax.scatter(coordinates[i, 0], coordinates[i, 1], coordinates[i, 2],
               color=colors[nucleotide], marker='o')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Helix")
plt.show()
```
