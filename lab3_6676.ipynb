```python
from collections import deque
def getneighbour(node,directions):
  arr = []
  for dir in directions:
    c =(0,0)
    c = list(c)
    c[0]=dir[0]+node[0]
    c[1]=dir[1]+node[1]
    c = tuple(c)
    if(not(c[0]>3 or c[0]<0 or c[1]>3 or c[1]<0)):
      arr.append(c)
  return arr

def shortest_path():
    start = (0, 0)
    end = (3, 3)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

    paths = [[start]]
    index = 0
    visited = {start}

    if start == end:
        return paths[0]

    while index < len(paths):
        current_route =paths[index]
        current_position =current_route[-1]
        neighbors = getneighbour(current_position, directions)
        if end in neighbors:
            current_route.append(end)
            return current_route

        for neighbor in neighbors:
            if neighbor not in visited:
                new_path =current_route[:]
                new_path.append(neighbor)
                paths.append(new_path)
                visited.add(neighbor)
        index += 1
    return []

print(shortest_path())
```

```python
import time
directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
def state_to_tuple(state):
    nested_tuple = tuple(
        tuple(
            int(state[row * 3 + col])
            for col in range(3)
        )
        for row in range(3)
    )
    return nested_tuple

def tuple_to_state(matrix):
  result = []
  for row in matrix:
    for cell in row:
        result.append(str(cell))
  return ''.join(result)

def get_moves(matrix):
    moves = []
    x, y = next((i, j) for i in range(3) for j in range(3) if matrix[i][j] == 0)

    for pos_x, pos_y in directions :
        new_x, new_y = x + pos_x, y + pos_y
        if 0 <= new_x < 3 and 0 <= new_y < 3:
            new_matrix =[]
            for row in matrix:
                new_matrix.append(list(row))
            temp =new_matrix[x][y]
            new_matrix[x][y] = new_matrix[new_x][new_y]
            new_matrix[new_x][new_y] =temp
            new_tuple_matrix =[]
            for row in new_matrix:
                new_tuple_matrix.append(tuple(row))
            moves.append(tuple(new_tuple_matrix))

    return moves

def dfs(start_state, goal_state):
    stack = [(start_state, [])]
    visited = {start_state}

    while stack:
        current_state, path = stack.pop()
        if current_state == goal_state:
            return path + [current_state]

        for next_state in get_moves(current_state):
            if next_state not in visited:
                visited.add(next_state)
                stack.append((next_state, path + [current_state]))

    return None

def main():
    start_state = input("Enter start state: ")
    goal_state = input("Enter goal state: ")

    start_tuple = state_to_tuple(start_state)
    goal_tuple = state_to_tuple(goal_state)
    print("\n-------------")
    print("\nDFS Algorithim")
    print("\n-------------")
    start_time = time.time()
    solution_path = dfs(start_tuple, goal_tuple)
    end_time = time.time()

    if solution_path:
        print(f"Time taken: {end_time - start_time:.4f} seconds")
        print(f"Path Cost: {len(solution_path) - 1}")
        print(f"Nodes Visited: {len(solution_path)}")

        for state in solution_path:
            for row in state:
                print(' '.join(map(str, row)))
            print("-----")
    else:
        print("No Sol")

if __name__ == "__main__":
    main()
```

```python
from collections import deque

class Graph:
    def __init__(self, adjacency_list):
        """Initializes the graph with an adjacency list."""
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        """Returns the neighbors of a given node."""
        return self.adjacency_list.get(v, [])

    def h(self, n):
        """Heuristic function: estimates the cost from node n to the goal."""
        H = {"The": 4, "cat": 3, "dog": 3, "runs": 2, "fast": 1}
        return H.get(n, float('inf'))

    def a_star_algorithm(self, start_node, stop_node):
        """Implements the A* search algorithm to find the optimal path."""
        open_list = set([start_node])
        closed_list = set([])

        g = {start_node: 0}  # Cost from start node to all other nodes
        f = {start_node: g[start_node] + self.h(start_node)}  # f(n) = g(n) + h(n)
        parents = {start_node: start_node}  # Keeps track of paths

        while len(open_list) > 0:
            n = None  # Current node

            # Find the node with the lowest f(n) = g(n) + h(n)
            for node in open_list:
                if n is None or f[node] < f[n]:
                    n =node

            if n is None:
                print("Path does not exist!")
                return None

            # If goal node is found, reconstruct and return the path
            if n == stop_node:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n=parents[n]
                reconst_path.append(start_node)
                reconst_path.reverse()
                print("Path Found:", ', '.join(reconst_path))
                return reconst_path

            # Move current node from open to closed list
            open_list.remove(n)
            closed_list.add(n)

            # Explore all neighbors of the current node
            for neighbor in self.get_neighbors(n):
                if neighbor in closed_list:
                    continue

                temp= g[n] + 1

                if neighbor not in open_list:
                    open_list.add(neighbor)
                elif temp >= g.get(neighbor, float('inf')):
                    continue

                # Update parent, g-score, and f-score f=g+h
                parents[neighbor] =n
                g[neighbor] =temp
                f[neighbor] =g[neighbor] + self.h(neighbor)

        print("No Path")
        return None

# Define the graph given in adjacency list
adjacency_list = { "The": ["cat", "dog"], "cat": ["runs"],"dog": ["runs"],"runs": ["fast"],"fast": []}
graph1 = Graph(adjacency_list)
graph1.a_star_algorithm("The", "fast")  # Define start and goal words
```
