```python
import time

class PuzzleState:
    def __init__(self, board, zero_pos, path_cost=0, path=[]):
        self.board = board
        self.zero_pos = zero_pos
        self.path_cost = path_cost
        self.path = path

    def get_possible_moves(self):
        moves = []
        x, y = self.zero_pos
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

        for dx, dy in directions:
            new_x, new_y = x + dx, y + dy
            if 0 <= new_x < 3 and 0 <= new_y < 3:
                new_board = [list(row) for row in self.board]

                new_board[x][y], new_board[new_x][new_y] = new_board[new_x][new_y], new_board[x][y]
                moves.append(PuzzleState(new_board, (new_x, new_y), self.path_cost + 1, self.path + [new_board]))

        return moves

    def is_goal(self, goal):
        return self.board == goal

def dfs(initial_state, goal_state):
    stack = [initial_state]
    visited = set()
    node_count = 0

    while stack:
        current_state = stack.pop()
        node_count += 1

        if current_state.is_goal(goal_state):
            return current_state.path_cost, node_count, current_state.path

        visited.add(tuple(map(tuple, current_state.board)))

        for next_state in current_state.get_possible_moves():
            if tuple(map(tuple, next_state.board)) not in visited:
                stack.append(next_state)

    return None, node_count, []

def print_board(board):
    for row in board:
        print(" ".join(str(x) for x in row))
    print("------")

def main():
    start_state_input = input("Enter start State: ")
    goal_state_input = input("Enter goal State: ")


    start_state = [list(map(int, start_state_input[i:i + 3])) for i in range(0, 9, 3)]
    goal_state = [list(map(int, goal_state_input[i:i + 3])) for i in range(0, 9, 3)]

    zero_pos = [(i, row.index(0)) for i, row in enumerate(start_state) if 0 in row][0]

    initial_puzzle = PuzzleState(start_state, zero_pos)

    start_time = time.time()
    path_cost, node_count, solution_path = dfs(initial_puzzle, goal_state)
    end_time = time.time()

    print("-----------------")
    print("DFS Algorithm")
    print("-----------------")
    print(f"Time taken: {end_time - start_time} seconds")
    print(f"Path Cost: {path_cost}")
    print(f"No of Node Visited: {node_count}")
    print("-----------------")

    for step in solution_path:
        print_board(step)

if __name__ == "__main__":
    main()
```

```python
from collections import defaultdict

class Find_PATH:
    def __init__(self, graph_structure):
        self.graph_structure = graph_structure

    def find_neighbors(self, node):
        return self.graph_structure[node]

    def heuristic(self, node):
        heuristic_values = {
            "fast": 1,
            "cat": 3,
            "dog": 3,
            "The": 4,
            "runs": 2
        }
        return heuristic_values[node]

    def A_star(self, start, goal):
        open_set = set([start])
        ClosedSeet = set([])

        costFromStr = {}
        costFromStr[start] = 0

        predecessors = {}
        predecessors[start] = None

        while len(open_set) > 0:
            current_node = None
            for node in open_set:
                if current_node is None or costFromStr[node] + self.heuristic(node) < costFromStr[current_node] + self.heuristic(current_node):
                    current_node = node

            if current_node is None:
                print("Path does not exist!")
                return None

            if current_node == goal:
                path = []
                while current_node is not None:
                    path.append(current_node)
                    current_node = predecessors[current_node]
                path.reverse()

                print("Total cost:", costFromStr[goal])
                print("Sentence:", " -> ".join(path))
                return path

            for (neighbor, travel_cost) in self.find_neighbors(current_node):
                if neighbor not in open_set and neighbor not in ClosedSeet:
                    open_set.add(neighbor)
                    predecessors[neighbor] = current_node
                    costFromStr[neighbor] = costFromStr[current_node] + travel_cost
                else:
                    if costFromStr[neighbor] > costFromStr[current_node] + travel_cost:
                        costFromStr[neighbor] = costFromStr[current_node] + travel_cost
                        predecessors[neighbor] = current_node

                        if neighbor in ClosedSeet:
                            ClosedSeet.remove(neighbor)
                            open_set.add(neighbor)

            open_set.remove(current_node)
            ClosedSeet.add(current_node)

        print("Path does not exist!")
        return None


graph_structure = {
    "The": [("cat", 1), ("dog", 2)],
    "cat": [("runs", 2)],
    "dog": [("runs", 2)],
    "runs": [("fast", 1)],
    "fast": []
}

Find_PATH = Find_PATH(graph_structure)
Find_PATH.A_star("The", "fast")
```
